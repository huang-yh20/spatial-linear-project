```python
import numpy
import matplotlib.pyplot as plt
```

```python

# 设置图形大小和分辨率
fig, ax = plt.subplots(figsize=(8 / 2.54, 8 / 2.54), dpi=300)

# 设置坐标轴范围
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])


# 生成坐标网格
grid_size = 10
x_E_list, y_E_list = np.meshgrid(np.linspace(0, 1, grid_size+1), np.linspace(0, 1, grid_size+1))
x_I_list, y_I_list = np.meshgrid(np.linspace(0, 1, grid_size//2+1), np.linspace(0, 1, grid_size//2+1))


ax.scatter(x_E_list, y_E_list, 
           s=25, marker='^', 
           color='#FF69B4',          # 粉色
           label='E',
           edgecolor='k', 
           linewidth=0.5,
           zorder=2)

ax.scatter(x_I_list, y_I_list-0.01,
           s=25, marker='o',
           color='#1F77B4',          # 蓝色
           label='I',
           edgecolor='k',
           linewidth=0.5,
           zorder=2)


# 设置边框
ax.set_frame_on(True)
ax.spines['top'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)

# 添加图例
plt.legend(loc='upper right')

# 保存图像
plt.savefig("../../figs/artfigs_NC_Fig1.svg", format='svg', bbox_inches='tight')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mu = 2  
sigma = 1 


x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)  
y = norm.pdf(x, mu, sigma)  


plt.figure(figsize=(8, 6))
# plt.plot(x, y, color='#8CC7B5', linewidth=4) 
plt.plot(x, y, color='b', linewidth=4) 

plt.axvline(x=2, ymin=0, ymax=1, color='gray', linestyle='--', linewidth=4)


ax = plt.gca()  
ax.set_xticks([])  
ax.set_yticks([])  


ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_linewidth(4)
ax.spines['left'].set_linewidth(4)


plt.ylim(bottom=0)


plt.xlabel(r'Synaptic Weight $J^{\alpha\beta}_{ij}$', fontsize=40)
plt.ylabel('Probability', fontsize=40)

# 去掉网格线
plt.grid(False)

# 显示图形
plt.tight_layout()
plt.savefig("../../figs/artfigs_NC_Fig1_conn_illus1.svg", format='svg', bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mu = 0
sigma = 0.1


x = np.linspace(-0.5, 0.5, 100)  
y = norm.pdf(x, mu, sigma)  

plt.figure(figsize=(8, 6))
# plt.axvline(x=0.5, ymin=0, ymax=1/(np.sqrt(2*np.pi) * 0.1), color='gray', linestyle='--', linewidth=4)
plt.axhline(y=1/(np.sqrt(2*np.pi) * 0.1), xmin=0, xmax=0.5, color='gray', linestyle='--', linewidth=4)
# plt.plot(x, y, color='#F4606C', linewidth=4) 
plt.plot(x, y, color='b', linewidth=4) 


ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_linewidth(4)
ax.spines['left'].set_linewidth(4)
ax.set_xticks([-0.5,0,0.5])  
ax.set_yticks([])  
ax.tick_params(axis='x', labelsize=40) 


plt.xlim(-0.5,0.5)
plt.ylim(0,5)


plt.xlabel('Distance', fontsize=40)
plt.ylabel(r'Conn. Prob. $P^{\alpha\beta}_{ij}$', fontsize=40)

plt.grid(False)

# 显示图形
plt.tight_layout()
plt.savefig("../../figs/artfigs_NC_Fig1_conn_illus2.svg", format='svg', bbox_inches='tight')
```
